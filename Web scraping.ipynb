```python
import requests
import csv
from bs4 import BeautifulSoup

def scrape_credit_card_details():
    url = "https://www.hdfcbank.com/personal/pay/cards/credit-cards"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    card_list = soup.find_all("div", class_="card-offer-contr")
    card_list1 = soup.find_all("div", class_="offer-dtl mB10")
    div = soup.find_all("div", class_="btnParent")
   
    data = []
    
    for card, card1,d in zip(card_list, card_list1,div):
        
        a=d.find("a",class_='btn btn-primary btn-custom btn-primary-customs hide-for-small mybtn normal-url',title="KNOW MORE")
        if a:
            anch=a.get("href")
            card_fee_url = "https://www.hdfcbank.com"+str(anch)+"/fees-and-charges"  # Replace with the URL for fetching card fees dynamically
            card_fee = fetch_card_fee(card_fee_url) 
           
        card_name = card.find("span", class_="card-name").text.strip()

        card_details = card1.find("ul").find_all("li")
       
       
       
        rewards=""
        lounge_access = "No"  # Assuming no lounge access by default
        milestone_benefit = "None"
        for i in card_details:
            rewards = rewards+i.text
            if "Lounge" in i.text:
                lounge_access = "Yes" 
            if "Milestone"in i.text or "milestone" in i.text:
                 milestone_benefit = "Yes"
                
            
        
        
       
        
        data.append([card_name, card_fee, rewards, lounge_access, milestone_benefit])
    
    return data
def fetch_card_fee(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    card_fee = ""
    
    inner_content = soup.find("div", class_="inner-content col-lg-8 col-sm-8 right-section")
    
    paragraph = inner_content.find("p")
    if "Fee" in paragraph.get_text(strip=True):
        card_fee = paragraph.get_text(strip=True)
    else:
        list_item = inner_content.find("ul")
        if list_item:
            list_item = list_item.find("li")
            if list_item and "Fee" in list_item.get_text(strip=True):
                card_fee = list_item.get_text(strip=True)

    return card_fee

def export_to_csv(data, filename):
    headers = ["Card Name", "Card fee", "Reward points/percentage per 100 spent", "Lounge access", "Milestone benefit"]
    
    with open(filename, "w", newline="", encoding="utf-8-sig") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(headers)
        writer.writerows(data)

# Scrape the data
card_data = scrape_credit_card_details()

# Export the data to a CSV file
export_to_csv(card_data, "credit_card_details.csv")
```
